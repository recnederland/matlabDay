## Veri Görselleştirme Nedir? 
<a id="optparam"></a>

Veri görselleştirme, karmaşık ve dağınık verileri düzenleyerek kolay anlaşılabilir, yorumlanabilir hale getirmektir. Bir başka ifade ile veri görselleştirme, elimizdeki soyut verilerin görsel hale getirilerek somutlaştırılması ve bundan bir ön bilgi elde edilmesi işidir.

**Veri görselleştirilmesi yapılırken dikkat edilmesi gereken bazı hususlar şunlardır;**

***-Renklerin uyumlu olması(bunun için renk paletleri kullanılabilir:***  http://colorbrewer2.org/#type=sequential&scheme=GnBu&n=4)

***-Sade ve yalın olması, fazlalıklara yer verilmemesi***

***-Doğru grafik türünün seçilmesi( Hangi veriyi hangi grafikle görselleştireceğiniz oldukça önemlidir.Bunun için veriyi tanımak ve grafik türlerini bilmek gerekir. Data görselleştirme kataloklarından yararlanılabilinir.***  https://datavizcatalogue.com/TR/)

***-Grafiğin anlaşılabilir olması***

***-Rakamların yerleşiminin doğru yapılması***


## Matplotlib Nedir?
<a id="optparam"></a>

Matplotlib; veri görselleştirmesinde kullandığımız temel python kütüphanesidir. 2 ve 3 boyutlu çizimler yapmamızı sağlar. Matplotlib genelde 2 boyutlu çizimlerde kullanılırken, 3 boyutlu çizimlerde başka kütüphanelerden yararlanılır.

## Kütüphanelerin Yüklenmesi ve Verilerin Düzenlenmesi
<a id="optparam"></a>

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

#verilerin yüklenmesi
data=pd.read_excel("matplot.xlsx")

#veri tipinin floata çevrilmesi
data['enflasyon'] = data['enflasyon'].astype('float')
data['issizlik'] = data['issizlik'].astype('float')

#ilk 5 verinin gösterilmesi
data.head()
````

````python
data.info()
````

## 1.Temel Görselleştirme ve Genel Komutlar
<a id="optparam"></a>

### 1.1.Görselleştirmede Genel Komutlar
<a id="optparam"></a>

Daha önceki derslerde belittiğim gibi kütüphaneleri yüklerken çeşitli kalıplaşmış kısaltmalar kullanılır. **```plt```**  kısaltması da matplotlib kütüphanesinin kısaltması olarak kullanılır. Grafikler bu komut ile çizilir.

**```plt.plot()```**: Grafiğin ana komutunu oluşturur.Bu komutun içine ilk önce x ekseni,sonrasında y ekseni yazılır. Grafikteki temel bazı değişkenleri değiştirmek istersek ilgili kodu buraya yazmamız gerekir. Örneğin grafiğin rengi, çizgi kalınlığı, aralıklar vs.

**``figsize=``**Grafiğin boyutunu belirler.**`f=plt.figure(figsize=(8,6))`** şeklinde yaygın olarak kullanılır. Farklı kullanımları da mevcuttur. Parantez içindeki ilk yazılan değer grafiğin eni, ikincisi boyudur.

**```plt.xlabel()```**: x ekseninin başlığını oluşturur.

**```plt.ylabel()```**: y ekseninin başlığını oluşturur.

**```plt.title()```**: Grafiğin başlığını oluşturur.

**```plt.show()```**: Grafiği çizdirmemize yarar. Bu kodu yazmazsak grafik çizilir fakat bazen ekranda görülmeyebilir.

**Bu komutları teker teker grafik üzerinde gösterelim.**

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon) #biz belirlemediğimiz sürece rengi otomatik kendisi verir. 

plt.show()
````

### 1.1.1.Grafiğe Başlık Eklemek
<a id="optparam"></a>

**```plt.title()```**: Grafiğin başlığını oluşturur.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon) #biz belirlemediğimiz sürece rengi otomatik kendisi verir. 

plt.title("Türkiye'nin Enflasyon Oranı")

plt.show()
````

### 1.1.2. X Eksenine Başlık Eklemek
<a id="optparam"></a>
**```plt.xlabel()```**: x ekseninin başlığını oluşturur.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon) #biz belirlemediğimiz sürece rengi otomatik kendisi verir. 

plt.title("Türkiye'nin Enflasyon Oranı")

plt.xlabel("Tarih")

plt.show()
````

### 1.1.3. Y Eksenine Başlık Eklemek
<a id="optparam"></a>

**```plt.ylabel()```**: y ekseninin başlığını oluşturur.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon) #biz belirlemediğimiz sürece rengi otomatik kendisi verir. 

plt.title("Türkiye'nin Enflasyon Oranı")

plt.xlabel("Tarih")

plt.ylabel("Enflasyon Oranları")
plt.show()
````

### 1.1.4. İki farklı veriyi tek grafikte gösterme
<a id="optparam"></a>

Tek bir grafikte iki farklı veriyi gösterebiliriz ve bu değişkenleri özelleştirebiliriz. Örneğin enflasyon ve işsizlik değişkenlerime faklı renkler veriyorum.
İkinci ya da daha fazla verinin görselleştirme kodunu eklerken ilk yazdığımız görselleştirme kodu gibi yazmamız yeterli olacaktır. **Burada dikkat edilmesi gereken nokta değişkenlerin x veya y eksenlerinin ortak olması gerekliliğidir.** Örneğimizde tarih iki değişken için ortaktır ve x ekseninde gösterilmiştir.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="red") 

plt.plot(data.Tarih,data.issizlik,color="blue")

plt.xlabel("Tarih")

plt.ylabel("Enflasyon ve İşsizlik Oranı")

plt.title("Türkiye'nin Enflasyon ve İşsizlik Oranları")

plt.show()
````

### 1.1.5.Birden fazla grafiği aynı anda çizme-subplot
<a id="optparam"></a>

**```subplot(x,y,z)```**: Grafiklerin düzlemini ve kaçıncı grafik olduğunu belirtir. İlk sayı satırı, ikinci sayı sütunu, üçüncü sayı ise kaçıncı grafik olduğunu ifade eder.

````python
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)   
plt.plot(data.Tarih,data.enflasyon,color="r") 
plt.xlabel("Tarih")
plt.ylabel("Enflasyon Oranı")
plt.title("Türkiye'nin Enflasyon Oranları")

plt.subplot(2,2,2)
plt.plot(data.Tarih,data.issizlik,color="blue")
plt.xlabel("Tarih")
plt.ylabel("İşsizlik Oranı")
plt.title("Türkiye'nin İşsizlik Oranları")

plt.show()
````

### 1.2.Figure Kullanarak Grafik Çizme
<a id="optparam"></a>

**``plt.figure()``**: Bu komut ile bir figür tanımlıyoruz. Yani aslında bir tablo içine bir şeyler yazmaya, çizmeye çalışıyoruz.

**``f.add_axes``**:Oluşturduğum figüre akses ekleyip koordinatlarını belirliyorum. **Burada yazdığım ilk değer soldan ne kadar içerde başlayacağı,ikinci değer alttan ne kadar yüksekte olacağı,üçüncü değer ne kadar uzunlukta olması gerektiği,dördüncü değer ise ne kadar yükseklikte olması gerektiğini belirtmektedir.** 0 ile 1 arasında değer almalıdır. Bu değerleri istediğiniz şekilde kendinize uygun formatta değiştirebilirsiniz.

````python
f=plt.figure(figsize=(10,4))

axes=f.add_axes([0.1,0.1,0.9,0.9])

axes.plot(data.Tarih,data.enflasyon)

axes.set_xlabel("Tarih")

axes.set_ylabel("Enflasyon")

axes.set_title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.2.1.İç İçe Grafik Çizme
<a id="optparam"></a>
Bazen grafikteki değerleri daha iyi anlayabilmek ya da karşılaştırabilmek için iç içe görmek gereklidir. Böyle bir durumda 2.axes oluşturmamız gereklidir.**Burada dikkat edilmesi gereken nokta grafiklerimizin koordinatlarıdır.** Grafiklerin görünürlüğü açısından verilen koordinatların yeri oldukça önemlidir.

````python
f=plt.figure(figsize=(12,4)) 


#koordinatlar
axes1=f.add_axes([0.1,0.1,0.9,0.9]) 

axes2=f.add_axes([0.6,0.3,0.3,0.3])


#verilerin grafikte nasıl yer alacağı.
axes1.plot(data.Tarih,data.enflasyon,color="blue")

axes2.plot(data.Tarih,data.issizlik,color="red")


#x ekseni başlıkları
axes1.set_xlabel("Tarih")
axes2.set_xlabel("Tarih")

#y ekseni başlıkları
axes1.set_ylabel("Enflasyon")
axes2.set_ylabel("işsizlik")

#Grafik başlıkları
axes1.set_title("Türkiye'nin Enflasyon Oranları")
axes2.set_title("Türkiye'nin İşsizlik Oranları")

plt.show()
````

### 1.3.Grafik Özelleştirme
<a id="optparam"></a>

### 1.3.1.Genel Komutlar
<a id="optparam"></a>

**`plt()`** ifadesinin aldığı komutlardır. Grafik üzerinde çeşitli değişiklikler yapılmasını sağlar.

**``color="red"``**: Grafiğin rengini belirler. Farklı renk paletleri yazılabilir.

**``linewidth=1``**: Grafikteki çizginin kalınlığını belirler.

**``linestyle="--"``**: Grafikteki çizginin stilini belirler.

**``marker="o"``**: Grafikteki her kırılım noktasına belirlediğimiz şekili koyar.

**``markersize=20``**: Marker'ın boyutunu belirler.

**``markerfacecolor="yellow"``**: Marker'ın rengini değiştirir.

**``markeredgewidth=4``**: Marker çizgisini kalınlaştırır.

**``markeredgecolor="blue"``**: Marker çizgisinin dış rengini değiştirir.

**``alpha=0.9``**: Saydamlığını değiştirir. 0'a yaklaştıkça saydamlaşır.1'e yaklaştıkça belirginleşir.

**Bu komutları grafikte teker teker uygulayalım.**

### 1.3.2.Grafiğin rengini belirleme
<a id="optparam"></a>

**``color``**: Grafiğin rengini belirler. Farklı renk paletleri yazılabilir.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="green") 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.3.Grafik çizgisinin kalınlığını belirleme
<a id="optparam"></a>

**``linewidth``**: Grafikteki çizginin kalınlığını belirler.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="r",linewidth=5) 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.4.Grafik çizgisinin stilini belirleme
<a id="optparam"></a>

**``linestyle``**: Grafikteki çizginin stilini belirler.


***linestyles*** = ['-',  '--',  '-.',  ':'] bu stilleri alır.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="r",linewidth=4,linestyle="--") 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.5.Grafik çizgisine marker çizme
<a id="optparam"></a>

**``marker="o"``**: Grafikteki her kırılım noktasına belirlediğimiz şekili koyar.

***Marker Stilleri:*** https://matplotlib.org/3.1.0/api/markers_api.html

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="r",linewidth=2,linestyle="--",marker="o") 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.6.Marker boyutunu belirleme
<a id="optparam"></a>


**``markersize``**: Marker boyutunu belirler.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="r",linewidth=3,linestyle="--",marker="o",markersize=10) 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.7.Marker rengini belirleme
<a id="optparam"></a>

**``markerfacecolor``**: Markerın rengini değiştirir.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,data.enflasyon,color="r",linewidth=3,linestyle="--",marker="o",markersize=12,markerfacecolor="yellow") 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.8.Marker kalınlığını belirleme
<a id="optparam"></a>

**``markeredgewidth``**: Marker çizgisini kalınlaştırır.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,
         data.enflasyon,
         color="r",
         linewidth=3,
         linestyle="--",
         marker="o",
         markersize=12,
         markerfacecolor="yellow",
         markeredgewidth=5) 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.9.Marker çizgisinin dış rengini belirleme
<a id="optparam"></a>

**``markeredgecolor``**: Marker çizgisinin dış rengini değiştirir.

````python
plt.figure(figsize=(12,6))

plt.plot(data.Tarih,
         data.enflasyon,
         color="r",
         linewidth=3,
         linestyle="--",
         marker="o",
         markersize=12,
         markerfacecolor="yellow",
         markeredgewidth=5,
         markeredgecolor="blue") 

plt.xlabel("Tarih")

plt.ylabel("Enflayon Oranları")

plt.title("Türkiye'nin Enflasyon Oranları")

plt.show()
````

### 1.3.10.Grafiğin saydamlığını belirleme
<a id="optparam"></a>

### 1.6.Grafiğe yazı yazma (text) komutları
<a id="optparam"></a>

**`plt.text`**: Grafiğe yazı eklememizi sağlar.

**`x,y`**: Yazacağımız yazının x ve y eksenine göre koordinatlarını belirtmemiz gerekir. ***``plt.text(3,4)``***

**`s`**: Yazacağımız yazıyı belirtmemiz  gerekir.***``plt.text(x=3,y=4,s="Bu yılda kriz olmuştur.")``***

**``fontdict``**: Yazacağımız yazı ile ilgili düzeltmeler yapmamızı sağlar. Sözlük formatında kullanılır. ***`plt.text(x=7,y=15,s="Bu yılda kriz olmuştur.", fontdict={"color":"r","fontsize":35})`*** 

**`withdash`**: True ve False değerler alır. Eklediğimiz yazının önüne çizgi çekmemizi sağlar.
